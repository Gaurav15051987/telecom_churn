```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
train_df=pd.read_csv(r'churn-bigml-80.csv')
test_df=pd.read_csv(r'churn-bigml-20.csv')
```

```python
train_df['Entire_Day_Mintue']=train_df['Total day minutes']+train_df['Total eve minutes']+train_df['Total night minutes']
train_df['Entire_Day_Call']=train_df['Total day calls']+train_df['Total eve calls']+train_df['Total night calls']
train_df['Entire_Day_charge']=train_df['Total day charge']+train_df['Total eve charge']+train_df['Total night charge']
train_df=train_df.drop(["State","Area code","Total day minutes","Total day calls","Total day charge","Total eve minutes","Total eve calls","Total eve charge","Total night minutes","Total night calls","Total night charge"],axis=1)
```

```python
cn=0
cn1=0
cn2=0
cn3=0
def get_numeric(data):
    cn=0
    for i in train_df.columns:
        if train_df[i].dtype!='O':
            cn=cn+1
    return cn
    
def get_char(data):
    cn=0
    for i in train_df.columns:
        if train_df[i].dtype=='O':
            cn=cn+1
    return cn
    
def get_bool(data):
    cn=0
    for i in train_df.columns:
        if train_df[i].dtype=='bool':
            cn=cn+1
    return cn
        
def get_factor(data):
    cn=0
    for i in train_df.columns:
        if train_df[i].dtype=='category':
            cn=cn+1
    return cn
```

```python
print("Total Numeric Columns is :", get_numeric(train_df))
print("Total Charcater Columns is :", get_char(train_df))
print("Total Boolen Columns is :", get_bool(train_df))
print("Total Factor Columns is :", get_factor(train_df))
```

```python
num_column=train_df._get_numeric_data()
df=pd.DataFrame()
for col in num_column.columns:
    df.loc[col,'Negative Value Count']=num_column[num_column[col]<0].shape[0]
    df.loc[col,'Postive Value Count']=num_column[num_column[col]>0].shape[0]
    df.loc[col,'Total Value Count']=num_column[col].count()
    df.loc[col,'Missing Value Count']=num_column[num_column[col].isna()].shape[0]
    df.loc[col,'Min Value Count']=num_column[col].min()
    df.loc[col,'Max Value Count']=num_column[col].max()
    df.loc[col,'Unique Value Count']=len(num_column[col].unique())
    df.loc[col,'Skw Value Count']=num_column[col].skew()
    df.loc[col,'Kut Value Count']=num_column[col].kurt()
```

```python
train_df['Churn']=np.where(train_df['Churn']==False,0,1) # False 0(churn No) True 1 (churn Yes)
train_df['International plan']=np.where(train_df['International plan']=='Yes',0,1) # False 0(churn No) True 1 (churn Yes)
train_df['Voice mail plan']=np.where(train_df['Voice mail plan']=='Yes',0,1) # False 0(churn No) True 1 (churn Yes)

# train_df['Churn'].value_counts()
train_df['International plan'].value_counts()
```

```python
features = train_df.drop(columns=['Churn']).columns
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=888)
X_resampled, y_resampled = smote.fit_resample(train_df[features], train_df['Churn'])

y_resampled.value_counts()
```

```python
X_resampled
```

```python
X=X_resampled.values
y=y_resampled.values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_selection import SelectKBest,chi2
```

```python
trf1=ColumnTransformer(transformers=[
    ('cnvt_lable',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,2]),
],remainder='passthrough')
# dd=pd.DataFrame(trf1.fit_transform(X_train))
# dd.info()
```

```python
trf2=ColumnTransformer(transformers=[
    ("scaling",MinMaxScaler(),slice(0,10)),
])
```

```python
trf3 = SelectKBest(score_func=chi2,k=8)
```

```python
trf4=LogisticRegression()
```

```python
pipe = Pipeline([
    ('cnvt_lable', trf1),
    ('scaling', trf2),
    ('best_feature', trf3),
    ('model_logistic', trf4)

])
```

```python
pipe = make_pipeline(trf1,trf2,trf3,trf4)
```

```python
from sklearn import set_config
set_config(display='diagram')
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps
```

```python
y_pred = pipe.predict(X_test)
y_pred_train = pipe.predict(X_train)
```

```python
y_pred = pipe.predict(X_test)
y_pred_train = pipe.predict(X_train)
from sklearn.metrics import accuracy_score
print("Train Accuracy",accuracy_score(y_train,y_pred_train))
print("Test Accuracy",accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
# pipeline_log_reg=Pipeline([
#     ('ohe',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,2]),
#     ('scaler',MinMaxScaler(),slice(0,10)),
#     ('best_feature', SelectKBest(score_func=chi2,k=8)),
#     ('logistic_reg',LogisticRegression())
# ])

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
pipeline_lr=Pipeline(steps=[
                    ('ohe1',OneHotEncoder(sparse=False,handle_unknown='ignore')),
                    ('MinMax1',MinMaxScaler()),
                     ('lr_classifier',LogisticRegression())
])


pipeline_DT=Pipeline(steps=[
                    ('ohe2',OneHotEncoder(sparse=False,handle_unknown='ignore')),
                    ('MinMax2',MinMaxScaler()),
#                      ('pca1',PCA(n_components=2)),
                     ('DT_classifier',DecisionTreeClassifier())
])


pipeline_SVM=Pipeline(steps=[
                    ('ohe3',OneHotEncoder(sparse=False,handle_unknown='ignore')),
                    ('MinMax3',MinMaxScaler()),
#                      ('pca1',PCA(n_components=2)),
                     ('SVM_classifier',svm.SVC(kernel='linear'))
])



pipeline_RF=Pipeline(steps=[
                    ('ohe4',OneHotEncoder(sparse=False,handle_unknown='ignore')),
                    ('MinMax4',MinMaxScaler()),
#                      ('pca1',PCA(n_components=2)),
                     ('RF_classifier',RandomForestClassifier())
])

pipeline_list=[pipeline_lr,pipeline_DT,pipeline_SVM,pipeline_RF]
```

```python
# from sklearn import set_config
# set_config(display='diagram')

pipeline_lr.fit(X_train, y_train)
pipeline_DT.fit(X_train, y_train)
pipeline_SVM.fit(X_train, y_train)
pipeline_RF.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
pipe_dict={0:'pipeline_lr',1:'pipeline_DT',2:'pipeline_SVM',3:'pipeline_RF'}
for i ,model in enumerate(pipeline_list):
    print("Accuary Score"+" "+pipe_dict[i]+":",model.score(X_test,y_test))
    print("Cross validation"+" "+pipe_dict[i]+":",cross_val_score(model,X_train,y_train,cv=5,scoring='accuracy').mean())
```

```python
for pipe2 in pipeline_list:
    print(sorted(pipe2.get_params().keys()))
```

```python
solver = ['newton-cg', 'lbfgs', 'liblinear','sag','saga']
penalty = ['l1', 'l2', 'elasticnet']
classifier= np.logspace(0, 4, 10)
max_iter=[1,2,4,5,None]
# class_weight=['balanced']
class_weight=['balanced']
# define grid search
grid_LR = dict(lr_classifier__solver=solver,lr_classifier__penalty=penalty,lr_classifier__C=classifier,
           lr_classifier__max_iter=max_iter,lr_classifier__class_weight=class_weight)



max_depth= [2, 3, 5, 10, 20]
min_samples_leaf=[5, 10, 20, 50, 100]
criterion=["gini", "entropy"]
grid_DT = dict(DT_classifier__max_depth=max_depth,DT_classifier__min_samples_leaf=min_samples_leaf
               ,DT_classifier__criterion=criterion,DT_classifier__class_weight=class_weight)



C=[0.1, 1, 10, 100, 1000]
gamma=[1, 0.1, 0.01, 0.001, 0.0001]
kernel= ['rbf']
grid_SVM = dict(SVM_classifier__C=C,SVM_classifier__gamma=gamma
               ,SVM_classifier__kernel=kernel,SVM_classifier__class_weight=class_weight)

n_estimators = [100,300,500,1000] # number of trees in the random forest
max_features = ['auto', 'sqrt'] # number of features in consideration at every split
max_depth = [1,3,5,7,10]# maximum number of levels allowed in each decision tree
min_samples_split = [2, 6, 10] # minimum sample number to split a node
min_samples_leaf = [1, 3, 4] # minimum sample number that can be stored in a leaf node

grid_RF = dict(RF_classifier__n_estimators=n_estimators,RF_classifier__max_features=max_features,RF_classifier__max_depth=max_depth,
               RF_classifier__min_samples_split=min_samples_split,RF_classifier__min_samples_leaf=min_samples_leaf)
params_list=['grid_LR','grid_DT','grid_SVM','grid_RF']
```

```python
from sklearn.model_selection import GridSearchCV
pipe_dict={0:'grid_LR',1:'grid_DT',2:'grid_SVM',3:'grid_RF'}

grid_LR1=GridSearchCV(pipeline_lr, grid_LR, cv=3, scoring='accuracy',verbose=True, n_jobs=-1)
grid_LR1.fit(X_train,y_train)

grid_DT1=GridSearchCV(pipeline_DT, grid_DT, cv=3, scoring='accuracy',verbose=True, n_jobs=-1)
grid_DT1.fit(X_train,y_train)

grid_SVM1=GridSearchCV(pipeline_SVM, grid_SVM, cv=3, scoring='accuracy',verbose=True, n_jobs=-1)
grid_SVM1.fit(X_train,y_train)

```

```python
grid_RF1=GridSearchCV(pipeline_RF, grid_RF, cv=3, scoring='accuracy',verbose=True, n_jobs=-1)
grid_RF1.fit(X_train,y_train)
```

```python
print("The mean accuracy of the Logistic model is:",grid_LR1.score(X_test,y_test))
print("The mean accuracy of Decision Tree the model is:",grid_DT1.score(X_test,y_test))
print("The mean accuracy of the SVM model is:",grid_SVM1.score(X_test,y_test))
# print("The mean accuracy of the Random Foreset model is:",grid_RF1.score(X_test,y_test))
```
